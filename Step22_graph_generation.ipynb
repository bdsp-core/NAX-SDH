```python
import numpy as np
import matplotlib.pyplot as plt

# Data
categories = ['ICDs', 'CPTs', 'Keywords', 'All Combined']
auroc = [0.9152, 0.7188, 0.9829, 0.9791]
auprc = [0.6595, 0.4810, 0.9327, 0.8961]
recall = [0.9507, 0.6203, 0.9739, 0.8406]
f1_score = [0.7558, 0.4477, 0.8842, 0.8555]

# Bar width
bar_width = 0.2

# Set positions of bars on X-axis
r1 = np.arange(len(categories))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]
r4 = [x + bar_width for x in r3]

# Define shades of blue
auroc_color = '#a6cee3'  # Light blue
auprc_color = '#1f77b4'  # Base blue
recall_color = '#004b87'  # Dark blue
f1_score_color = '#003f5c'  # Darker blue

# Create the bar plot
plt.figure(figsize=(12, 7))

bars_auroc = plt.bar(r1, auroc, color=auroc_color, width=bar_width, edgecolor='grey', label='AUROC')
bars_auprc = plt.bar(r2, auprc, color=auprc_color, width=bar_width, edgecolor='grey', label='AUPRC')
bars_recall = plt.bar(r3, recall, color=recall_color, width=bar_width, edgecolor='grey', label='Recall')
bars_f1_score = plt.bar(r4, f1_score, color=f1_score_color, width=bar_width, edgecolor='grey', label='F1 Score')

# Add values on top of the bars
def add_values(bars):
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width() / 2, height, f'{height:.2f}', ha='center', va='bottom')

add_values(bars_auroc)
add_values(bars_auprc)
add_values(bars_recall)
add_values(bars_f1_score)

# Add xticks on the middle of the group bars with category names
plt.xticks([r + bar_width*1.5 for r in range(len(categories))], categories)

# Hide the x-axis label "Categories"
plt.xlabel('')  # Empty string to hide the x-axis label

# Create the legend with a larger font size
plt.legend(fontsize=12, loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=4, frameon=False)

plt.title('Performance Metrics Comparison')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
categories = ['ICDs', 'CPTs', 'Keywords', 'All Combined']
auroc = [0.9152, 0.7188, 0.9829, 0.9791]
auprc = [0.6595, 0.4810, 0.9327, 0.8961]
recall = [0.9507, 0.6203, 0.9739, 0.8406]
f1_score = [0.7558, 0.4477, 0.8842, 0.8555]

# Prepare data for the table
table_data = [auroc, auprc, recall, f1_score]
column_labels = ['ICDs', 'CPTs', 'Keywords', 'All Combined']
row_labels = ['AUROC', 'AUPRC', 'Recall', 'F1 Score']
cell_text = np.array(table_data).tolist()

# Create the figure and axis
plt.figure(figsize=(12, 6))

# Create the table
plt.table(cellText=cell_text,
          rowLabels=row_labels,
          colLabels=column_labels,
          cellLoc='center',
          loc='center',
          bbox=[0, 0, 1, 1])  # Adjust bbox for positioning

plt.title('Performance Metrics Table')
plt.axis('off')  # Hide the axis

plt.tight_layout()
plt.show()
```
