```python
#Import all of the necessary modules. 
import numpy as np
import matplotlib.pyplot as plt
```

```python
#This code creates a blue version of the figure containing RF tests for comparision. 

#Data for RF models.
categories = ['RF All Combined', 'RF Keywords Only', 'RF CPT Codes Only', 'RF ICD Codes Only']
auroc = [0.979, 0.985, 0.719, 0.915]
auprc = [0.896, 0.944, 0.481, 0.660]
recall = [0.841, 0.971, 0.620, 0.986]
f1_score = [0.856, 0.890, 0.448, 0.761]

#Confidence intervals for RF models
auroc_ci = [(0.973, 0.985), (0.981, 0.990), (0.692, 0.743), (0.902, 0.929)]
auprc_ci = [(0.851, 0.934), (0.924, 0.962), (0.438, 0.520), (0.602, 0.717)]
recall_ci = [(0.804, 0.876), (0.951, 0.987), (0.569, 0.668), (0.971, 0.997)]
f1_score_ci = [(0.827, 0.881), (0.865, 0.913), (0.406, 0.486), (0.727, 0.793)]

#Calculate error bars
auroc_err = [((high - low) / 2) for low, high in auroc_ci]
auprc_err = [((high - low) / 2) for low, high in auprc_ci]
recall_err = [((high - low) / 2) for low, high in recall_ci]
f1_score_err = [((high - low) / 2) for low, high in f1_score_ci]

#Bar width and space between groups
bar_width = 0.3
group_spacing = 0.3  

#Set positions of bars on X-axis
r1 = np.arange(len(categories)) * (bar_width * 4 + group_spacing) 
r2 = [x + bar_width for x in r1]  
r3 = [x + bar_width for x in r2]  
r4 = [x + bar_width for x in r3] 

#Define shades of blue
auroc_color = '#a6cee3'  
auprc_color = '#1f77b4'  
recall_color = '#004b87'  
f1_score_color = '#003f5c'  

#Create the bar plot for RF models
plt.figure(figsize=(14, 7))  
bars_auroc = plt.bar(r1, auroc, yerr=auroc_err, color=auroc_color, width=bar_width, edgecolor='grey', label='AUROC', capsize=5)
bars_auprc = plt.bar(r2, auprc, yerr=auprc_err, color=auprc_color, width=bar_width, edgecolor='grey', label='AUPRC', capsize=5)
bars_recall = plt.bar(r3, recall, yerr=recall_err, color=recall_color, width=bar_width, edgecolor='grey', label='Recall', capsize=5)
bars_f1_score = plt.bar(r4, f1_score, yerr=f1_score_err, color=f1_score_color, width=bar_width, edgecolor='grey', label='F1 Score', capsize=5)

#Add values above the error bars
def add_values_with_error_bars(bars, errors):
    for bar, err in zip(bars, errors):
        height = bar.get_height()
        error_height = err
        plt.text(bar.get_x() + bar.get_width() / 2, height + error_height + 0.02, 
                 f'{height:.3f}', ha='center', va='bottom', fontsize=14)

add_values_with_error_bars(bars_auroc, auroc_err)
add_values_with_error_bars(bars_auprc, auprc_err)
add_values_with_error_bars(bars_recall, recall_err)
add_values_with_error_bars(bars_f1_score, f1_score_err)

#Add xticks on the middle of the group bars with category names
plt.xticks([r + (bar_width * 1.5) for r in r1], categories, fontsize=14)

#Set the y-axis label font size
plt.yticks(fontsize=14)

#Hide the x-axis label "Categories"
plt.xlabel('', fontsize=14) 

#Create the legend with a larger font size
plt.legend(fontsize=14, loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=4, frameon=False)

#Hide top and right spines
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.tight_layout()
plt.show()
```
